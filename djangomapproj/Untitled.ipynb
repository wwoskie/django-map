```python
import folium
import pandas as pd
import geopandas as gpd
```

```python
table = gpd.read_file('table_after.csv', sep='\t')
```

```python
table
```

```python
table = gpd.GeoDataFrame(table)
```

```python
table
```

```python
m = folium.Map(zoom_start=5, 
                              tiles='cartodbpositron', 
                              location = [55.799534, 37.460218], 
                              prefer_canvas=True,
                              max_bounds=True)

folium.Choropleth(
        geo_data=table['geometry'],
    ).add_to(m)
```
